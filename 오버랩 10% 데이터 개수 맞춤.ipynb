```python
import scipy.io.wavfile
import scipy.signal
import numpy as np
import heartpy as hp
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
from tqdm import tqdm
import os
from sklearn.model_selection import train_test_split
```

```python
## train, test 데이터 분리(train 70, test 30) 및 한 파일로 병합(overlap 10%)

ECG_train = pd.DataFrame()
ECG_test = pd.DataFrame()
alltrain = []
alltest = []
for k in range(1,101):
    data_url = 'D:/Data/ECGDB_preprocessing/07_1. overlap(10%)/' + str(k)
    sub_dir = os.listdir(data_url)
    train, test = train_test_split(sub_dir, test_size=0.3, shuffle=True, random_state=34)    
    
    for i in train:
        file = data_url + '/' + i
        dataset = pd.read_csv(file, header=0)
        
        dataset.columns = ['No','time','signal{0}'.format(k)]
        alltrain.append(dataset['signal{0}'.format(k)])
        ECG_train = pd.concat(alltrain, axis=1, ignore_index=False)
    

    for i in test:
        file = data_url + '/' + i
        dataset = pd.read_csv(file, header=0)
        
        dataset.columns = ['No','time','signal{0}'.format(k)]
        alltest.append(dataset['signal{0}'.format(k)])
        ECG_test = pd.concat(alltest, axis=1, ignore_index=False)
    
    
print(ECG_train)
print(ECG_test)
```
